```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import random

work_days = {
    '2024-01': 24,
    '2024-02':23,
    '2024-03':21,
    '2024-04':24,
    '2024-05':23,
    '2024-06':20,
    '2024-07':24,
    '2024-08':22,
    '2024-09':23,
    '2024-10':24,
    '2024-11':22,
    '2024-12':23
}
```

```python
x = pd.read_excel('./cajas_revisadas ENERO 2024 A ENERO 2025.xlsx')
x.head()
df = x[['CONCATENAR', 'cd_usua',]]
df['fecha'] =  [str(i)[0:7] for i in x['fe_actu']]
df_2024 = df[df['fecha'].astype(str).str.contains('2024')]
df_2024
```

```python
names = set(name for name in df_2024['cd_usua'])
```

```python
def num_p(name):
    return len(df_2024[df_2024['cd_usua'] == name])

box_name_year = {i:num_p(i) for i in names}


box_name_month = {month:set(df_2024['cd_usua'][df_2024['fecha'] == month]) for month in work_days}

num_box_year = {}
for month in df_2024['fecha']:
    if month in num_box_year:
        num_box_year[month] += 1
    else:
        num_box_year[month] = 1


def total_productivity_day(date):

    return round(num_box_year[date]/(len(box_name_month[date]))/(work_days[date]),2)
    #return round(num_box_year[date]/(5)/(work_days[date]),2)




def total_productivity_hour(date):
    response = (num_box_year[date]/len(box_name_month[date]))/(work_days[date]*8)
    return round(response, 2)
```

```python
prod_2024_import_day = {month:total_productivity_day(date=month) for month in work_days}
prod_2024_import_hour = {month:total_productivity_hour(date=month) for month in work_days}


num_pers_month = [len(box_name_month[i]) for i in box_name_month ]

prod_2024 = pd.DataFrame(
    {'Productividad por día': prod_2024_import_day.values(), 
    'Productividad por hora': prod_2024_import_hour.values(),
    'Número de personas': num_pers_month},
    index=prod_2024_import_day.keys()
    )
prod_2024

```
